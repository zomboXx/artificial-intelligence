Mục tiêu duy nhất là tìm ra **trạng thái đích (Goal State)** hoặc trạng thái tốt nhất có thể, thường ứng dụng cho các bài toán tối ưu hóa (Optimization Problems) như xếp thời khóa biểu, 8-Queens.

## 2. Thuật toán IDA* (Iterative Deepening A*)
### a. Ý tưởng và Quy trình hoạt động
**IDA*** là sự kết hợp hoàn hảo giữa cơ chế "sâu dần" của **IDS** và hàm đánh giá heuristic của **A***. 
- Thay vì giới hạn độ sâu (Depth Limit) như IDS, IDA* sử dụng **giới hạn chi phí (Cost Limit)** dựa trên hàm $f(n) = g(n) + h(n)$.
- **Quy trình**:
  1. Khởi tạo giới hạn $threshold = f(Start)$.
  2. Duyệt DFS từ Start. Bất kỳ nhánh nào có $f(n) > threshold$, ta sẽ cắt nhánh (prune) và bỏ qua.
  3. Nếu tìm thấy Goal, thuật toán kết thúc.
  4. Nếu duyệt hết mà chưa thấy Goal, cập nhật $threshold$ mới bằng giá trị $f(n)$ nhỏ nhất trong số các nút vừa bị cắt, rồi lặp lại bước 2.

### b. Đánh giá 4 đặc tính của thuật toán
1. **Tính đầy đủ (Completeness)**: **CÓ**. Miễn là chi phí mỗi bước đi $c \ge \epsilon > 0$ và Heuristic không đánh giá vô cực sai lệch, thuật toán sẽ vét cạn đến khi chạm Goal.
2. **Tính tối ưu (Optimality)**: **CÓ**. Tương tự A*, IDA* đảm bảo tính tối ưu miễn là hàm $h(n)$ là *Admissible* (Chấp nhận được - không bao giờ đánh giá khoảng cách lớn hơn thực tế).
3. **Độ phức tạp không gian (Space Complexity)**: $\mathbf{O(b \cdot d)}$. **Đây là "Vũ khí tối thượng" của IDA***. Vì bản chất duyệt theo DFS, nó chỉ cần lưu nhánh từ Root đến Current Node. Không cần tập `Frontier` hay `Explored` khổng lồ như A*. 
4. **Độ phức tạp thời gian (Time Complexity)**: Phụ thuộc vào số lượng giá trị Heuristic khác nhau. Worst-case là $O(b^d)$ hoặc có thể tệ hơn nếu mỗi lần lặp ngưỡng chỉ tăng lên 1 đơn vị rất nhỏ. Tuy nhiên, trong thực tế, chi phí sinh lại các node ở phần ngọn cây chiếm tỷ lệ cực nhỏ so với sự bùng nổ tổ hợp ở đáy cây. Do đó, thời gian dư thừa là cái giá hoàn toàn xứng đáng để đổi lấy bộ nhớ $O(b \cdot d)$.

### c. Đánh giá Hàm Heuristic (Cost Function)
- IDA* cực kỳ nhạy cảm với số lượng các giá trị $f(n)$ khác nhau. 
- Trong bài toán 8-Puzzle, dùng **Manhattan Distance** thường tốt hơn **Misplaced Tiles** vì Manhattan đưa ra ước lượng sát với chi phí thực tế hơn, giúp cắt tỉa (prune) không gian tìm kiếm mạnh mẽ hơn.

### d. So sánh: IDA* vs A*
- **A***: Chạy nhanh hơn nhưng "ngốn" RAM khủng khiếp (lưu toàn bộ Frontier và Explored). Thường sẽ bị văng (Out of Memory) với các bài toán có không gian trạng thái lớn (như 15-Puzzle).
- **IDA***: Giải quyết triệt để bài toán bộ nhớ của A*, đánh đổi bằng việc chậm hơn một chút do phải duyệt lại (Iterative). Với các game/bài toán lớn, IDA* là lựa chọn thực tế hơn A*.

## 3. Thuật toán Tìm kiếm cục bộ - Hill Climbing (Leo đồi)
### a. Ý tưởng và Quy trình hoạt động
Trong Local Search, ta tưởng tượng không gian trạng thái như một địa hình đồi núi:
- **Trục x, y**: Biểu diễn không gian các trạng thái có thể có (State space).
- **Trục z (Độ cao)**: Biểu diễn **Hàm mục tiêu (Objective Function)** hoặc **Hàm Heuristic**. 
  - Nếu ta muốn tối đa hóa điểm số $\rightarrow$ Bài toán **Leo đồi (Hill Climbing)**. Đích là đỉnh cao nhất (*Global Maximum*).
  - Nếu ta muốn tối thiểu hóa số lỗi (như số cặp hậu cắn nhau trong 8-Queens) $\rightarrow$ Bài toán **Xuống lũng (Gradient Descent)**. Đích là điểm thấp nhất (*Global Minimum*). Hai dạng này có chung bản chất toán học.

- **Quy trình**: Bắt đầu từ một trạng thái ngẫu nhiên, tạo ra các trạng thái kề (Neighbors). Nếu neighbor tốt hơn trạng thái hiện tại (hàm mục tiêu cao hơn hoặc chi phí thấp hơn), di chuyển sang đó. Thuật toán dừng lại khi không có neighbor nào tốt hơn.

### b. Phân loại
1. **Simple Hill Climbing (Leo đồi đơn giản)**: Đánh giá lần lượt các neighbor. Ngay khi tìm thấy một neighbor **đầu tiên** tốt hơn trạng thái hiện tại, lập tức di chuyển đến đó luôn. (Nhanh, nhưng dễ đi sai hướng).
2. **Steepest-Ascent Hill Climbing (Leo đồi dốc nhất)**: Đánh giá **tất cả** các neighbor. Sau đó chọn ra neighbor **tốt nhất (độ dốc cao nhất)** để di chuyển. (Tốn thời gian đánh giá hơn nhưng bước đi chất lượng hơn).

### c. Đánh giá 4 đặc tính & Các vấn đề cạm bẫy
1. **Độ phức tạp không gian (Space Complexity)**: $\mathbf{O(b)}$ hoặc $\mathbf{O(1)}$. Thuật toán chỉ lưu duy nhất *Current Node* và (tối đa) danh sách *Neighbors* của nó. Giải phóng hoàn toàn RAM so với mọi thuật toán tìm kiếm truyền thống.
2. **Độ phức tạp thời gian (Time Complexity)**: Phụ thuộc địa hình. Worst-case là $O(\infty)$ nếu bị kẹt trong vòng lặp vô hạn ở địa hình bằng phẳng (nếu không cấu hình điều kiện dừng hợp lý).
3. **Tính đầy đủ (Completeness)**: **KHÔNG**. Thuật toán vô cùng dễ bị kẹt và không bao giờ đến đích.
4. **Tính tối ưu (Optimality)**: **KHÔNG**. Thuật toán mang tính "cận thị" (short-sighted), nó dừng lại ngay khi chạm tới một đỉnh (dù là đỉnh thấp), không hề có cái nhìn toàn cục để biết đâu mới thực sự là cấu hình tốt nhất.

5. **Đi sâu vào 3 cạm bẫy kinh điển (Tại sao Hill Climbing thất bại?)**
+ **Local Maxima (Cực đại cục bộ)**: 
   - *Bản chất*: Là một đỉnh cao hơn tất cả các neighbors kề cạnh nó, nhưng lại thấp hơn đỉnh cao nhất thực sự (Global Maximum).
   - *Biểu hiện*: Thuật toán kẹt cứng tại đây vì mọi bước đi tiếp theo đều dẫn đến giá trị giảm xuống (mà nguyên lý của leo đồi là không bao giờ đi lùi).
+ **Plateau / Flat Local Maximum (Cao nguyên / Bình nguyên)**:
   - *Bản chất*: Một vùng không gian bằng phẳng, nơi mọi trạng thái lân cận đều có hàm mục tiêu bằng hệt như trạng thái hiện tại.
   - *Biểu hiện*: Thuật toán mất phương hướng. Nó có thể chọn di chuyển ngẫu nhiên và rơi vào vòng lặp vô tận (infinite loop) quanh quẩn trên cao nguyên đó mà không tìm thấy dốc để leo.
+ **Ridge (Dãy núi / Sống núi)**:
   - *Bản chất*: Một vùng có độ dốc cao nhưng hẹp. Hướng dốc nhất lại không nằm trong các hướng di chuyển hợp lệ (các toán tử lân cận không cho phép đi chéo theo sống núi).
   - *Biểu hiện*: Thuật toán liên tục thực hiện các bước di chuyển zigzag qua lại hai bên sườn núi. Điều này làm thuật toán hội tụ cực kỳ chậm và rất dễ dừng lại sớm trước khi tới đỉnh.

### d. So sánh & Khắc phục
- So với các thuật toán tìm kiếm mù hay A*, Hill Climbing code cực dễ, bộ nhớ siêu thấp, phù hợp cho bài toán siêu lớn (như tối ưu trọng số Neural Network cơ bản, hay bài 8-Queens).
- Để khắc phục nhược điểm mắc kẹt của Hill Climbing, ta thường dùng biến thể **Random-restart Hill Climbing** (Chạy lại từ đầu với một điểm Start ngẫu nhiên nhiều lần. Nếu chạy vô hạn lần, tính Completeness sẽ tiến tới 1), hoặc kết hợp với **Simulated Annealing (Luyện kim)** (Thỉnh thoảng cho phép thuật toán "đi lùi" (đi xuống thung lũng) để có cơ hội nhảy thoát ra khỏi Local Maxima. Xác suất đi lùi này sẽ giảm dần theo thời gian.)

## 4. Hướng dẫn chạy & Đánh giá trên Source Code
1. Mở file Notebook tương ứng. Chạy các cell cài đặt hai thuật toán **IDA*** và **Hill Climbing**.
2. **Với IDA***: Hãy thử thay đổi trọng số hoặc cấu hình Heuristic (Manhattan so với Misplaced Tiles) để quan sát việc: Khi Heuristic đánh giá kém, số vòng lặp cập nhật `threshold` tăng vọt, kéo theo lượng node `Generated` tăng chóng mặt như thế nào.
3. **Với Hill Climbing**: 
   - Thử nghiệm trên bài toán 8-Queens (8 quân hậu). Hàm $h(n)$ lúc này là số cặp hậu cắn nhau (cần đưa về 0).
   - Trong quá trình chạy nhiều test case ngẫu nhiên, hãy in ra terminal trạng thái khi vòng lặp dừng lại. Bạn sẽ thấy rõ tỷ lệ thất bại (kẹt ở $h(n) > 0$) là cực kỳ cao, chứng minh trực quan tính "KHÔNG Completeness" của thuật toán này.

```python
# =========================================================================
# CHƯƠNG TRÌNH MÔ PHỎNG: IDA* VÀ LOCAL SEARCH (HILL CLIMBING)
# - Hỗ trợ chọn bài toán: 8-Puzzle (Mặc định) và 8-Queens.
# - Hỗ trợ chọn thuật toán tùy theo bài toán.
# - Giao diện tự động thích ứng vẽ ma trận 3x3 hoặc bàn cờ 8x8.
# - Hiển thị rõ các trạng thái mắc kẹt (Local Maxima/Plateau).
# - Tự động tối ưu giao diện 8-Queens bằng thẻ Text để chống lag.
# =========================================================================

import tkinter as tk
from tkinter import ttk

# --- CẤU HÌNH TRẠNG THÁI ---
PUZZLE_START = ((1, 2, 3), (4, 0, 6), (7, 5, 8))
PUZZLE_GOAL  = ((1, 2, 3), (4, 5, 6), (7, 8, 0))
# 8-Queens: Khởi tạo một trạng thái xấu: Tất cả quân hậu nằm ở hàng 0 để xem thuật toán tối ưu dần.
QUEENS_START = (0, 0, 0, 0, 0, 0, 0, 0) 

MOVES = [(-1, 0, 'Up'), (1, 0, 'Down'), (0, -1, 'Left'), (0, 1, 'Right')]

# ── 1. HELPERS CHO 8-PUZZLE ───────────────────────────────
def get_neighbors_puzzle(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                x, y = i, j
    result = []
    for dx, dy, act in MOVES:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            s = [list(r) for r in state]
            s[x][y], s[nx][ny] = s[nx][ny], s[x][y]
            result.append((tuple(tuple(r) for r in s), act))
    return result

def manhattan_distance(state, goal):
    dist = 0
    goal_pos = {goal[r][c]: (r, c) for r in range(3) for c in range(3)}
    for r in range(3):
        for c in range(3):
            val = state[r][c]
            if val != 0:
                tr, tc = goal_pos[val]
                dist += abs(r - tr) + abs(c - tc)
    return dist

# ── 2. HELPERS CHO 8-QUEENS ───────────────────────────────
def h_8queens(state):
    # Đếm số cặp hậu tấn công lẫn nhau (cùng hàng hoặc cùng đường chéo)
    h = 0
    for i in range(8):
        for j in range(i + 1, 8):
            if state[i] == state[j] or abs(state[i] - state[j]) == j - i:
                h += 1
    return h

def get_neighbors_queens(state):
    neighbors = []
    for col in range(8):
        for row in range(8):
            if state[col] != row:
                new_state = list(state)
                new_state[col] = row
                neighbors.append((tuple(new_state), f'Col {col}➔Row {row}'))
    return neighbors

def reconstruct(node):
    path = []
    while node:
        path.append(node['state'])
        node = node['parent']
    return path[::-1]

# ── 3. THUẬT TOÁN: IDA* SEARCH ────────────────────────────
def ida_star_gen(start, goal):
    threshold = manhattan_distance(start, goal)
    expanded = 0
    generated = 0
    
    while True:
        stack = [({'state': start, 'parent': None, 'g': 0, 'h': threshold, 'f': threshold}, set([start]))]
        next_threshold = float('inf')
        
        while stack:
            node, current_path = stack.pop()
            state = node['state']
            f = node['f']
            
            frontier_nodes = [item[0] for item in stack]
            yield node, frontier_nodes, [], expanded, generated, None, f'IDA* (Threshold: {threshold})'
            
            if f > threshold:
                next_threshold = min(next_threshold, f)
                continue
                
            if state == goal:
                yield node, [], [], expanded, generated, reconstruct(node), 'IDA* SUCCESS'
                return
                
            expanded += 1
            for nb, _ in reversed(get_neighbors_puzzle(state)):
                if nb not in current_path:
                    nb_g = node['g'] + 1
                    nb_h = manhattan_distance(nb, goal)
                    nb_f = nb_g + nb_h
                    
                    new_path = current_path.copy()
                    new_path.add(nb)
                    stack.append(({'state': nb, 'parent': node, 'g': nb_g, 'h': nb_h, 'f': nb_f}, new_path))
                    generated += 1
                    
        if next_threshold == float('inf'):
            return 
        threshold = next_threshold

# ── 4. THUẬT TOÁN: HILL CLIMBING (ĐÃ NÂNG CẤP XANH/ĐỎ) ──────────────────
def hill_climbing_gen(start, problem_type, variant):
    current_state = start
    expanded = 0
    generated = 0
    
    def get_h(s): return manhattan_distance(s, PUZZLE_GOAL) if problem_type == '8-Puzzle' else h_8queens(s)
    def get_nbrs(s): return get_neighbors_puzzle(s) if problem_type == '8-Puzzle' else get_neighbors_queens(s)
    
    current_h = get_h(current_state)
    node = {'state': current_state, 'parent': None, 'h': current_h, 'f': current_h}
    
    yield node, [], [], expanded, generated, None, variant
    
    while True:
        if (problem_type == '8-Queens' and current_h == 0) or (problem_type == '8-Puzzle' and current_state == PUZZLE_GOAL):
            yield node, [], [], expanded, generated, reconstruct(node), f'{variant} SUCCESS'
            return
            
        expanded += 1
        raw_neighbors = get_nbrs(current_state)
        generated += len(raw_neighbors)
        
        # Đóng gói dữ liệu toàn bộ danh sách neighbors cục bộ phục vụ UI vẽ màu sắc
        frontier_neighbors = []
        for nbr, act in raw_neighbors:
            h_val = get_h(nbr)
            frontier_neighbors.append({
                'state': nbr,
                'h': h_val,
                'f': h_val,
                'action': act,
                'status': 'Rejected' # Mặc định ban đầu là bị loại
            })
            
        chosen_nbr_dict = None
        
        if variant == 'Simple Hill Climbing':
            # Chọn neighbor ĐẦU TIÊN tốt hơn current nút hiện tại
            for n in frontier_neighbors:
                if n['h'] < current_h:
                    n['status'] = 'Chosen'
                    chosen_nbr_dict = n
                    break
        else:
            # Steepest-Ascent: Tìm giá trị h nhỏ nhất trong toàn bộ tập
            min_h = min(n['h'] for n in frontier_neighbors)
            if min_h < current_h:
                for n in frontier_neighbors:
                    if n['h'] == min_h:
                        n['status'] = 'Chosen'
                        chosen_nbr_dict = n
                        break
                        
        # Gửi toàn bộ danh sách neighbors kèm trạng thái sang UI để vẽ màu Xanh/Đỏ
        yield node, frontier_neighbors, [], expanded, generated, None, variant
        
        # NẾU KHÔNG CÓ NEIGHBOR NÀO TỐT HƠN -> KẸT (LOCAL MAXIMA / PLATEAU)
        if chosen_nbr_dict is None:
            yield node, frontier_neighbors, [], expanded, generated, None, f'FAILED (Kẹt tại Local Maxima / Plateau)'
            return
            
        # Tiến lên trạng thái mới
        current_state = chosen_nbr_dict['state']
        current_h = chosen_nbr_dict['h']
        node = {'state': current_state, 'parent': node, 'h': current_h, 'f': current_h}

# ── Scrollable Frame ─────────────────────────────────────
class ScrollableFrame(tk.Frame):
    def __init__(self, parent, height=180):
        super().__init__(parent)
        self.h_scroll = tk.Scrollbar(self, orient='horizontal')
        self.h_scroll.pack(side='bottom', fill='x')
        self.canvas = tk.Canvas(self, height=height)
        self.canvas.pack(side='left', fill='both', expand=True)
        self.h_scroll.config(command=self.canvas.xview)
        self.canvas.configure(xscrollcommand=self.h_scroll.set)
        self.inner = tk.Frame(self.canvas)
        self.inner.bind('<Configure>', lambda e: self.canvas.configure(scrollregion=self.canvas.bbox('all')))
        self.canvas.create_window((0, 0), window=self.inner, anchor='nw')
        self.canvas.bind('<Shift-MouseWheel>', lambda e: self.canvas.xview_scroll(int(-1 * (e.delta / 120)), 'units'))

# ── GUI (ĐÃ TỐI ƯU CHỐNG LAG 8-QUEENS & ĐỔI MÀU XANH ĐỎ) ──────────────────
class PuzzleGUI:
    def __init__(self, root):
        self.root = root
        self.gen = None
        self.running = False
        self.step_count = 0
        
        root.title('AI Search – Multi-Problem Visualization')
        sw, sh = root.winfo_screenwidth(), root.winfo_screenheight()
        w, h = min(1400, int(sw * .85)), min(900, int(sh * .85))
        root.geometry(f'{w}x{h}+{(sw-w)//2}+{(sh-h)//2}')
        root.minsize(900, 600)

        self.info = tk.Label(root, text='Select Problem and Algorithm, then click Reset', font=('Arial', 14, 'bold'))
        self.info.pack(pady=6)

        bf = tk.Frame(root)
        bf.pack(side='bottom', fill='x', pady=8)
        center_f = tk.Frame(bf)
        center_f.pack(anchor='center')
        
        tk.Label(center_f, text='Problem:', font=('Arial', 10, 'bold')).pack(side='left', padx=(5, 2))
        self.prob_cb = ttk.Combobox(center_f, values=['8-Puzzle', '8-Queens'], state='readonly', width=12)
        self.prob_cb.set('8-Puzzle')
        self.prob_cb.pack(side='left', padx=5)
        self.prob_cb.bind('<<ComboboxSelected>>', self.update_algorithm_list)

        tk.Label(center_f, text='Algorithm:', font=('Arial', 10, 'bold')).pack(side='left', padx=(15, 2))
        self.algo_cb = ttk.Combobox(center_f, state='readonly', width=20)
        self.algo_cb.pack(side='left', padx=5)
        
        tk.Button(center_f, text='Reset', width=8, command=self.reset_search, bg='#FF5722', fg='white').pack(side='left', padx=(10,5))
        tk.Button(center_f, text='Next Step', width=12, command=self.next_step).pack(side='left', padx=5)
        self.auto_btn = tk.Button(center_f, text='Auto Run', width=12, command=self.toggle_auto)
        self.auto_btn.pack(side='left', padx=5)

        self.speed_var = tk.IntVar(value=300)
        tk.Label(center_f, text='Speed:').pack(side='left', padx=(15, 2))
        tk.Scale(center_f, from_=10, to=1000, orient='horizontal', variable=self.speed_var, showvalue=False, length=120).pack(side='left')

        self.cur_frame = tk.LabelFrame(root, text='Current State (Popped Node)', padx=8, pady=8)
        self.cur_frame.pack(side='top', fill='x', padx=10, pady=4)

        self.fl_frame = tk.LabelFrame(root, text='Frontier / Neighbors', padx=8, pady=8)
        self.fl_frame.pack(side='top', fill='both', expand=True, padx=10, pady=4)
        self.frontier_sf = ScrollableFrame(self.fl_frame, 220)
        self.frontier_sf.pack(fill='both', expand=True)
        
        self.update_algorithm_list(None)

    def update_algorithm_list(self, event):
        prob = self.prob_cb.get()
        if prob == '8-Puzzle':
            self.algo_cb['values'] = ['IDA* Search', 'Simple Hill Climbing', 'Steepest-Ascent HC']
            self.algo_cb.set('IDA* Search')
        else:
            self.algo_cb['values'] = ['Simple Hill Climbing', 'Steepest-Ascent HC']
            self.algo_cb.set('Simple Hill Climbing')
        self.reset_search()

    def reset_search(self):
        self.running = False
        self.step_count = 0
        self.auto_btn.config(text='Auto Run')
        self.info.config(text='Ready! Click Next Step to Start')
        self.clear(self.cur_frame)
        self.clear(self.frontier_sf.inner)
        
        prob = self.prob_cb.get()
        algo = self.algo_cb.get()
        
        start_state = PUZZLE_START if prob == '8-Puzzle' else QUEENS_START
        
        if algo == 'IDA* Search':
            self.gen = ida_star_gen(start_state, PUZZLE_GOAL)
            self.fl_frame.config(text='Frontier (Stack DFS) - Checked Threshold')
        elif algo == 'Simple Hill Climbing':
            self.gen = hill_climbing_gen(start_state, prob, 'Simple Hill Climbing')
            self.fl_frame.config(text='Immediate Neighbors Evaluated (Best/First: Green, Rejected: Red)')
        elif algo == 'Steepest-Ascent HC':
            self.gen = hill_climbing_gen(start_state, prob, 'Steepest-Ascent HC')
            self.fl_frame.config(text='Immediate Neighbors Evaluated (Best: Green, Rejected: Red)')
            
        self.draw_state(self.cur_frame, start_state, f'Initial Start State').pack()

    def draw_state(self, parent, state, label_text=None, bg_color="white"):
        container = tk.Frame(parent, bg=bg_color, bd=2, relief='groove')
        grid_f = tk.Frame(container, bg=bg_color)
        grid_f.pack(padx=4, pady=4)
        
        if isinstance(state[0], int): # 8-Queens: (0, 0, 0...)
            for r in range(8):
                for c in range(8):
                    bg_cell = '#FFFFFF' if (r+c)%2==0 else '#D3D3D3'
                    is_queen = (state[c] == r)
                    tk.Label(grid_f, text='♛' if is_queen else '', width=2, height=1,
                             font=('Arial', 12, 'bold'), relief='solid', borderwidth=1,
                             bg=bg_cell, fg='#FF0000' if is_queen else 'black').grid(row=r, column=c)
        else: # 8-Puzzle: ((1,2,3),...)
            for i in range(3):
                for j in range(3):
                    v = state[i][j]
                    tk.Label(grid_f, text=str(v) if v else '', width=4, height=2,
                             font=('Arial', 13, 'bold'), relief='solid',
                             bg='#4CAF50' if v == 0 else 'white').grid(row=i, column=j)
                             
        if label_text:
            lbl = tk.Label(container, text=label_text, font=('Arial', 10, 'bold'), fg='black', bg=bg_color)
            lbl.pack(pady=4)
        return container

    def clear(self, frame):
        for w in frame.winfo_children(): w.destroy()

    def render(self, node, frontier, explored, expanded, generated, path, mode):
        self.clear(self.cur_frame)
        self.clear(self.frontier_sf.inner)

        c_state, c_h = node['state'], node['h']
        info_str = f"h(n) = {c_h}"
        if 'g' in node: info_str = f"f={node['f']} (g={node['g']}, h={c_h})"
        
        # Luôn vẽ trạng thái hiện tại (kể cả 8-Queens) dưới dạng bàn cờ đầy đủ cực kỳ đẹp mắt
        self.draw_state(self.cur_frame, c_state, info_str).pack()
        
        # Duyệt qua các neighbor
        for i, n in enumerate(frontier):
            n_state, n_h = n['state'], n['h']
            n_info = f"h(n) = {n_h}"
            if 'g' in n: n_info = f"f={n['f']} (g={n['g']}, h={n_h})"
            
            # Đọc trạng thái từ thuật toán để quyết định màu nền Xanh / Đỏ
            status = n.get('status', '')
            if status == 'Chosen':
                bg_c = '#C8E6C9' # Màu xanh ngọt
                n_info += "\n[CHOSEN]"
            elif status == 'Rejected':
                bg_c = '#FFCDD2' # Màu đỏ nhạt
                n_info += "\n[Rejected]"
            else:
                bg_c = 'white'
                
            if 'action' in n:
                n_info = f"{n['action']}\n" + n_info
                
            # ĐÂY LÀ ĐOẠN CỨU CÁNH HIỆU NĂNG CHO 8-QUEENS:
            if isinstance(n_state[0], int):
                # Thay vì sinh ra 56 bàn cờ con, ta vẽ một thẻ Text Box gọn nhẹ đầy đủ thông tin màu sắc
                card = tk.Frame(self.frontier_sf.inner, bg=bg_c, bd=2, relief='groove', padx=8, pady=8)
                tk.Label(card, text=n_info, font=('Arial', 9, 'bold'), bg=bg_c, justify='center').pack()
                card.grid(row=0, column=i, padx=5, pady=6)
            else:
                # 8-Puzzle ít trạng thái nên vẫn vẽ bàn cờ 3x3 như bình thường
                self.draw_state(self.frontier_sf.inner, n_state, n_info, bg_c).grid(row=0, column=i, padx=8, pady=8)

        status = f'[{mode}] Step: {self.step_count} | Expanded: {expanded} | Generated: {generated}'
        if path: status += f' | 🎉 REACHED GOAL (Path: {len(path)-1})'
        if 'FAILED' in mode: status += f' | ❌ THUẬT TOÁN BỊ KẸT!'
        self.info.config(text=status)

    def next_step(self):
        try:
            node, frontier, explored, exp, gen, path, mode = next(self.gen)
            self.step_count += 1
            self.render(node, frontier, explored, exp, gen, path, mode)
            if path or 'FAILED' in mode:
                self.running = False
                self.auto_btn.config(text='Auto Run')
        except StopIteration:
            self.running = False
            self.auto_btn.config(text='Done')

    def toggle_auto(self):
        self.running = not self.running
        self.auto_btn.config(text='Pause' if self.running else 'Auto Run')
        if self.running: self.auto_tick()

    def auto_tick(self):
        if not self.running: return
        self.next_step()
        if self.running: self.root.after(self.speed_var.get(), self.auto_tick)

root = tk.Tk()
app = PuzzleGUI(root)
root.mainloop()
```
