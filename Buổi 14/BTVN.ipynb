# Báo cáo bài tập về nhà Buổi 14: Tìm kiếm đối kháng

Trong buổi học này, tôi học nhóm thuật toán **tìm kiếm đối kháng**. Đây là nhóm thuật toán dùng trong các môi trường có nhiều tác nhân, đặc biệt là trò chơi hai người, trong đó hành động của một bên ảnh hưởng trực tiếp tới lợi ích của bên còn lại.

Nếu các buổi trước tôi thường xét một tác nhân đi từ start tới goal, thì ở tìm kiếm đối kháng, tôi phải xét thêm yếu tố **đối thủ**. Vì vậy, bài toán không chỉ là “tôi đi bước nào tốt nhất”, mà còn là “nếu tôi đi bước này, đối thủ sẽ phản ứng thế nào”.

Ba thuật toán chính trong notebook này là:

1. **Minimax**
2. **Alpha-Beta Pruning**
3. **Expectimax** - đây là thuật toán thứ ba nhiều khả năng tương ứng với phần tôi chưa nhớ tên trong buổi học. Nếu Minimax giả định đối thủ luôn tối ưu, thì Expectimax dùng khi một số node được xem là node xác suất.

## 1. Minimax

Minimax dùng cho trò chơi hai người, tổng lợi ích bằng không hoặc gần như đối lập nhau. Một bên là **MAX**, muốn tối đa hóa điểm số. Bên còn lại là **MIN**, muốn tối thiểu hóa điểm số của MAX.

Tư tưởng:

```text
Nếu đến lượt MAX: chọn nước đi có giá trị lớn nhất
Nếu đến lượt MIN: chọn nước đi có giá trị nhỏ nhất
Nếu gặp trạng thái kết thúc: trả về utility
```

Với Tic-Tac-Toe:

- MAX là `X`.
- MIN là `O`.
- Nếu `X` thắng, utility dương.
- Nếu `O` thắng, utility âm.
- Nếu hòa, utility bằng 0.

Minimax đầy đủ nếu cây game hữu hạn, nhưng có thể tốn nhiều thời gian vì phải xét nhiều nhánh.

## 2. Alpha-Beta Pruning

Alpha-Beta không thay đổi kết quả của Minimax, nhưng giảm số node cần xét bằng cách cắt bỏ các nhánh chắc chắn không thể ảnh hưởng tới quyết định cuối cùng.

Hai biến quan trọng:

- **Alpha**: giá trị tốt nhất mà MAX hiện có thể đảm bảo.
- **Beta**: giá trị tốt nhất mà MIN hiện có thể đảm bảo.

Nếu trong quá trình duyệt ta có:

```text
alpha >= beta
```

thì nhánh còn lại có thể bị cắt, vì đối thủ sẽ không cho phép ta đi tới nhánh đó.

Tư tưởng chính: **vẫn ra quyết định như Minimax, nhưng không xét những nhánh không còn ý nghĩa**.

## 3. Expectimax

Expectimax dùng khi một số node không còn là đối thủ tối ưu tuyệt đối, mà là **chance node** hoặc tác nhân hành động theo xác suất. Khi đó, thay vì lấy min tại lượt đối thủ, tôi lấy kỳ vọng trung bình của các kết quả.

Tư tưởng:

```text
Nếu đến lượt MAX: chọn giá trị lớn nhất
Nếu đến chance node: lấy giá trị kỳ vọng của các successor
Nếu gặp trạng thái kết thúc: trả về utility
```

Trong notebook này, tôi minh họa Expectimax bằng cách xem lượt `O` như một đối thủ ngẫu nhiên đều trên các nước đi hợp lệ. Vì vậy, giá trị của một nước đi không phải là trường hợp xấu nhất như Minimax, mà là giá trị trung bình kỳ vọng.

## 4. So sánh ba thuật toán

| Thuật toán | Mô hình đối thủ | Cách chọn giá trị | Kết quả |
|---|---|---|---|
| Minimax | Đối thủ tối ưu | MAX lấy max, MIN lấy min | Nước đi an toàn trước đối thủ giỏi |
| Alpha-Beta | Đối thủ tối ưu | Giống Minimax nhưng có cắt nhánh | Cùng kết quả với Minimax, thường nhanh hơn |
| Expectimax | Đối thủ/chance có xác suất | MAX lấy max, chance lấy kỳ vọng | Tốt khi môi trường hoặc đối thủ không hoàn toàn tối ưu |

## 5. Bài toán minh họa

Tôi dùng **Tic-Tac-Toe** để minh họa vì cây game nhỏ, dễ quan sát và thể hiện rõ vai trò của MAX/MIN/chance node.

Trong giao diện:

- `X` là AI/MAX.
- `O` là đối thủ.
- Có thể chọn thuật toán `Minimax`, `Alpha-Beta`, hoặc `Expectimax`.
- Mỗi lần AI đi, chương trình hiển thị các nước đi ứng viên, giá trị đánh giá, số node đã mở rộng và số nhánh bị cắt.

Điểm cần chú ý: `Expectimax` có thể chọn nước đi khác Minimax, vì nó không giả định `O` luôn đi nước tốt nhất để chống lại `X`.

## 6. Bổ sung trực quan hóa cây tìm kiếm

Để quan sát thuật toán rõ hơn, giao diện không chỉ hiển thị bàn cờ hiện tại mà còn hiển thị **root của cây tìm kiếm** và các **node con ứng với từng nước đi hợp lệ của MAX**. Mỗi node con có giá trị đánh giá. Node được chọn được tô nổi bật.

- Với **Minimax**, node con thể hiện giá trị khi `O/MIN` phản ứng tối ưu.
- Với **Alpha-Beta**, giá trị vẫn giống Minimax, nhưng giao diện hiển thị thêm số nhánh bị cắt và trace các thời điểm `alpha >= beta`.
- Với **Expectimax**, node con là giá trị kỳ vọng khi lượt `O` được xem như chance node.

Nhờ vậy, tôi có thể debug theo ba lớp: bàn cờ hiện tại, các nước đi ứng viên ở root, và trace quá trình duyệt/cắt nhánh.

## 7. Debug từng bước thuật toán

Để trực quan hóa đúng hơn, tôi bổ sung chế độ **debug từng bước**. Thay vì chỉ xem nước đi cuối cùng, tôi quan sát từng sự kiện trong quá trình đệ quy:

- `ENTER`: thuật toán đi xuống một node mới.
- `DESCEND`: thuật toán chuẩn bị xét một successor.
- `LEAF`: gặp trạng thái kết thúc hoặc đạt giới hạn độ sâu debug.
- `UPDATE`: node cha cập nhật giá trị tốt nhất hoặc giá trị kỳ vọng.
- `PRUNE`: Alpha-Beta cắt các nhánh còn lại khi `alpha >= beta`.
- `RETURN`: node hiện tại trả value về node cha.

Như vậy, phần debug không chỉ trả lời “AI chọn ô nào”, mà còn cho thấy thuật toán đã duyệt cây như thế nào, value được truyền ngược ra sao và nhánh nào bị bỏ qua.


````python
# =========================================================================
# BUỔI 14 - ADVERSARIAL SEARCH: MINIMAX, ALPHA-BETA, EXPECTIMAX
# Minh họa bằng Tic-Tac-Toe
# =========================================================================

import math
import random
import tkinter as tk
from tkinter import ttk


MAX_PLAYER = "X"
MIN_PLAYER = "O"
EMPTY = " "
WIN_LINES = (
    (0, 1, 2), (3, 4, 5), (6, 7, 8),
    (0, 3, 6), (1, 4, 7), (2, 5, 8),
    (0, 4, 8), (2, 4, 6),
)


def empty_board():
    return tuple(EMPTY for _ in range(9))


def board_after(board, move, player):
    next_board = list(board)
    next_board[move] = player
    return tuple(next_board)


def legal_moves(board):
    return [index for index, value in enumerate(board) if value == EMPTY]


def winner(board):
    for a, b, c in WIN_LINES:
        if board[a] != EMPTY and board[a] == board[b] == board[c]:
            return board[a]
    if EMPTY not in board:
        return "Draw"
    return None


def utility(board, depth=0):
    result = winner(board)
    if result == MAX_PLAYER:
        return 10 - depth
    if result == MIN_PLAYER:
        return depth - 10
    return 0


def opponent(player):
    return MIN_PLAYER if player == MAX_PLAYER else MAX_PLAYER


def move_label(move):
    row = move // 3 + 1
    col = move % 3 + 1
    return f"r{row}c{col}"


def add_trace(stats, text, limit=14):
    trace = stats.setdefault("trace", [])
    if len(trace) < limit:
        trace.append(text)


def node_role(player, algorithm):
    if algorithm == "Expectimax" and player == MIN_PLAYER:
        return "CHANCE"
    return "MAX" if player == MAX_PLAYER else "MIN"


def evaluate_with_algorithm(board, algorithm):
    if algorithm == "Minimax":
        return root_minimax(board)
    if algorithm == "Alpha-Beta":
        return root_alpha_beta(board)
    return root_expectimax(board)


def root_minimax(board):
    stats = {"nodes": 0, "pruned": 0, "trace": []}
    candidates = []
    add_trace(stats, "ROOT/MAX: xét các nước đi hợp lệ của X.")
    for move in legal_moves(board):
        value = minimax(board_after(board, move, MAX_PLAYER), MIN_PLAYER, 1, stats)
        candidates.append((move, value))
    candidates.sort(key=lambda item: (-item[1], item[0]))
    best_move, best_value = candidates[0] if candidates else (None, utility(board))
    return {
        "move": best_move,
        "value": best_value,
        "candidates": candidates,
        "nodes": stats["nodes"],
        "pruned": stats["pruned"],
        "message": "Minimax: MAX chọn nước đi có giá trị lớn nhất, giả định O luôn tối ưu.",
        "trace": stats["trace"],
    }


def minimax(board, player, depth, stats):
    stats["nodes"] += 1
    result = winner(board)
    if result is not None:
        return utility(board, depth)

    moves = legal_moves(board)
    if depth <= 2:
        add_trace(stats, f"{node_role(player, 'Minimax')} depth {depth}: xét {len(moves)} successor.")
    values = [
        minimax(board_after(board, move, player), opponent(player), depth + 1, stats)
        for move in moves
    ]
    return max(values) if player == MAX_PLAYER else min(values)


def root_alpha_beta(board):
    stats = {"nodes": 0, "pruned": 0, "trace": []}
    candidates = []
    add_trace(stats, "ROOT/MAX: duyệt với alpha=-inf, beta=+inf.")
    alpha = -math.inf
    beta = math.inf
    for move in legal_moves(board):
        value = alpha_beta(board_after(board, move, MAX_PLAYER), MIN_PLAYER, 1, alpha, beta, stats)
        candidates.append((move, value))
        alpha = max(alpha, value)
    candidates.sort(key=lambda item: (-item[1], item[0]))
    best_move, best_value = candidates[0] if candidates else (None, utility(board))
    return {
        "move": best_move,
        "value": best_value,
        "candidates": candidates,
        "nodes": stats["nodes"],
        "pruned": stats["pruned"],
        "message": "Alpha-Beta: kết quả như Minimax nhưng cắt các nhánh không còn ảnh hưởng.",
        "trace": stats["trace"],
    }


def alpha_beta(board, player, depth, alpha, beta, stats):
    stats["nodes"] += 1
    result = winner(board)
    if result is not None:
        return utility(board, depth)

    if player == MAX_PLAYER:
        value = -math.inf
        for move in legal_moves(board):
            value = max(value, alpha_beta(board_after(board, move, player), MIN_PLAYER, depth + 1, alpha, beta, stats))
            alpha = max(alpha, value)
            if alpha >= beta:
                stats["pruned"] += 1
                add_trace(stats, f"PRUNE depth {depth}: alpha={alpha:.3f} >= beta={beta:.3f}.")
                break
        return value

    value = math.inf
    for move in legal_moves(board):
        value = min(value, alpha_beta(board_after(board, move, player), MAX_PLAYER, depth + 1, alpha, beta, stats))
        beta = min(beta, value)
        if alpha >= beta:
            stats["pruned"] += 1
            add_trace(stats, f"PRUNE depth {depth}: alpha={alpha:.3f} >= beta={beta:.3f}.")
            break
    return value


def root_expectimax(board):
    stats = {"nodes": 0, "pruned": 0, "trace": []}
    candidates = []
    add_trace(stats, "ROOT/MAX: xét nước đi của X, sau đó O là chance node.")
    for move in legal_moves(board):
        value = expectimax(board_after(board, move, MAX_PLAYER), MIN_PLAYER, 1, stats)
        candidates.append((move, value))
    candidates.sort(key=lambda item: (-item[1], item[0]))
    best_move, best_value = candidates[0] if candidates else (None, utility(board))
    return {
        "move": best_move,
        "value": best_value,
        "candidates": candidates,
        "nodes": stats["nodes"],
        "pruned": stats["pruned"],
        "message": "Expectimax: lượt O được xem như chance node, lấy kỳ vọng thay vì lấy min.",
        "trace": stats["trace"],
    }


def expectimax(board, player, depth, stats):
    stats["nodes"] += 1
    result = winner(board)
    if result is not None:
        return utility(board, depth)

    moves = legal_moves(board)
    if depth <= 2:
        add_trace(stats, f"{node_role(player, 'Expectimax')} depth {depth}: xét {len(moves)} successor.")
    values = [
        expectimax(board_after(board, move, player), opponent(player), depth + 1, stats)
        for move in moves
    ]
    if player == MAX_PLAYER:
        return max(values)
    return sum(values) / len(values)




def heuristic_value(board):
    result = winner(board)
    if result is not None:
        return utility(board, 0)

    score = 0
    weights = {0: 0, 1: 1, 2: 4, 3: 10}
    for line in WIN_LINES:
        values = [board[index] for index in line]
        x_count = values.count(MAX_PLAYER)
        o_count = values.count(MIN_PLAYER)
        if o_count == 0:
            score += weights[x_count]
        if x_count == 0:
            score -= weights[o_count]
    return score


def format_path(path):
    return "ROOT" if not path else "ROOT -> " + " -> ".join(move_label(move) for move in path)


def short_board(board):
    return "".join("." if value == EMPTY else value for value in board)


def make_debug_event(phase, board, player, depth, path, message, **extra):
    event = {
        "phase": phase,
        "board": board,
        "player": player,
        "depth": depth,
        "path": list(path),
        "message": message,
    }
    event.update(extra)
    return event


def build_debug_events(board, algorithm, root_player=MAX_PLAYER, max_depth=3):
    if algorithm == "Alpha-Beta":
        return trace_alpha_beta_debug(board, root_player, max_depth)
    if algorithm == "Expectimax":
        return trace_expectimax_debug(board, root_player, max_depth)
    return trace_minimax_debug(board, root_player, max_depth)


def trace_minimax_debug(board, root_player=MAX_PLAYER, max_depth=3):
    events = []

    def recurse(state, player, depth, path):
        role = "MAX" if player == MAX_PLAYER else "MIN"
        events.append(make_debug_event(
            "ENTER", state, player, depth, path,
            f"Vào node {role} ở depth {depth}.",
            role=role,
        ))

        result = winner(state)
        if result is not None or depth >= max_depth:
            value = utility(state, depth) if result is not None else heuristic_value(state)
            reason = "terminal" if result is not None else "cutoff heuristic"
            events.append(make_debug_event(
                "LEAF", state, player, depth, path,
                f"Leaf ({reason}) trả value={value:.3f}.",
                role=role,
                value=value,
            ))
            return value

        child_values = []
        best = -math.inf if player == MAX_PLAYER else math.inf
        for move in legal_moves(state):
            child = board_after(state, move, player)
            events.append(make_debug_event(
                "DESCEND", state, player, depth, path,
                f"{role} xét child {move_label(move)}.",
                role=role,
                current_move=move,
                child_values=list(child_values),
            ))
            value = recurse(child, opponent(player), depth + 1, path + [move])
            child_values.append((move, value))
            best = max(best, value) if player == MAX_PLAYER else min(best, value)
            events.append(make_debug_event(
                "UPDATE", state, player, depth, path,
                f"{role} nhận {move_label(move)}={value:.3f}, best hiện tại={best:.3f}.",
                role=role,
                current_move=move,
                value=best,
                child_values=list(child_values),
            ))

        events.append(make_debug_event(
            "RETURN", state, player, depth, path,
            f"{role} trả value={best:.3f} về node cha.",
            role=role,
            value=best,
            child_values=list(child_values),
        ))
        return best

    recurse(board, root_player, 0, [])
    return events


def trace_alpha_beta_debug(board, root_player=MAX_PLAYER, max_depth=3):
    events = []

    def recurse(state, player, depth, path, alpha, beta):
        role = "MAX" if player == MAX_PLAYER else "MIN"
        events.append(make_debug_event(
            "ENTER", state, player, depth, path,
            f"Vào node {role}: alpha={alpha:.3f}, beta={beta:.3f}.",
            role=role,
            alpha=alpha,
            beta=beta,
        ))

        result = winner(state)
        if result is not None or depth >= max_depth:
            value = utility(state, depth) if result is not None else heuristic_value(state)
            reason = "terminal" if result is not None else "cutoff heuristic"
            events.append(make_debug_event(
                "LEAF", state, player, depth, path,
                f"Leaf ({reason}) trả value={value:.3f}.",
                role=role,
                alpha=alpha,
                beta=beta,
                value=value,
            ))
            return value

        child_values = []
        moves = legal_moves(state)
        if player == MAX_PLAYER:
            value = -math.inf
            for index, move in enumerate(moves):
                events.append(make_debug_event(
                    "DESCEND", state, player, depth, path,
                    f"MAX xét {move_label(move)} với alpha={alpha:.3f}, beta={beta:.3f}.",
                    role=role,
                    alpha=alpha,
                    beta=beta,
                    current_move=move,
                    child_values=list(child_values),
                ))
                child_value = recurse(board_after(state, move, player), MIN_PLAYER, depth + 1, path + [move], alpha, beta)
                child_values.append((move, child_value))
                value = max(value, child_value)
                alpha = max(alpha, value)
                events.append(make_debug_event(
                    "UPDATE", state, player, depth, path,
                    f"MAX cập nhật value={value:.3f}, alpha={alpha:.3f}.",
                    role=role,
                    alpha=alpha,
                    beta=beta,
                    value=value,
                    current_move=move,
                    child_values=list(child_values),
                ))
                if alpha >= beta:
                    pruned = moves[index + 1:]
                    events.append(make_debug_event(
                        "PRUNE", state, player, depth, path,
                        f"Cắt {len(pruned)} nhánh vì alpha={alpha:.3f} >= beta={beta:.3f}.",
                        role=role,
                        alpha=alpha,
                        beta=beta,
                        value=value,
                        pruned_moves=pruned,
                        child_values=list(child_values),
                    ))
                    break
            events.append(make_debug_event(
                "RETURN", state, player, depth, path,
                f"MAX trả value={value:.3f}.",
                role=role,
                alpha=alpha,
                beta=beta,
                value=value,
                child_values=list(child_values),
            ))
            return value

        value = math.inf
        for index, move in enumerate(moves):
            events.append(make_debug_event(
                "DESCEND", state, player, depth, path,
                f"MIN xét {move_label(move)} với alpha={alpha:.3f}, beta={beta:.3f}.",
                role=role,
                alpha=alpha,
                beta=beta,
                current_move=move,
                child_values=list(child_values),
            ))
            child_value = recurse(board_after(state, move, player), MAX_PLAYER, depth + 1, path + [move], alpha, beta)
            child_values.append((move, child_value))
            value = min(value, child_value)
            beta = min(beta, value)
            events.append(make_debug_event(
                "UPDATE", state, player, depth, path,
                f"MIN cập nhật value={value:.3f}, beta={beta:.3f}.",
                role=role,
                alpha=alpha,
                beta=beta,
                value=value,
                current_move=move,
                child_values=list(child_values),
            ))
            if alpha >= beta:
                pruned = moves[index + 1:]
                events.append(make_debug_event(
                    "PRUNE", state, player, depth, path,
                    f"Cắt {len(pruned)} nhánh vì alpha={alpha:.3f} >= beta={beta:.3f}.",
                    role=role,
                    alpha=alpha,
                    beta=beta,
                    value=value,
                    pruned_moves=pruned,
                    child_values=list(child_values),
                ))
                break
        events.append(make_debug_event(
            "RETURN", state, player, depth, path,
            f"MIN trả value={value:.3f}.",
            role=role,
            alpha=alpha,
            beta=beta,
            value=value,
            child_values=list(child_values),
        ))
        return value

    recurse(board, root_player, 0, [], -math.inf, math.inf)
    return events


def trace_expectimax_debug(board, root_player=MAX_PLAYER, max_depth=3):
    events = []

    def recurse(state, player, depth, path):
        role = "MAX" if player == MAX_PLAYER else "CHANCE"
        events.append(make_debug_event(
            "ENTER", state, player, depth, path,
            f"Vào node {role} ở depth {depth}.",
            role=role,
        ))

        result = winner(state)
        if result is not None or depth >= max_depth:
            value = utility(state, depth) if result is not None else heuristic_value(state)
            reason = "terminal" if result is not None else "cutoff heuristic"
            events.append(make_debug_event(
                "LEAF", state, player, depth, path,
                f"Leaf ({reason}) trả value={value:.3f}.",
                role=role,
                value=value,
            ))
            return value

        child_values = []
        moves = legal_moves(state)
        for move in moves:
            events.append(make_debug_event(
                "DESCEND", state, player, depth, path,
                f"{role} xét child {move_label(move)}.",
                role=role,
                current_move=move,
                child_values=list(child_values),
            ))
            value = recurse(board_after(state, move, player), opponent(player), depth + 1, path + [move])
            child_values.append((move, value))

            if player == MAX_PLAYER:
                aggregate = max(v for _, v in child_values)
                msg = f"MAX cập nhật best={aggregate:.3f}."
            else:
                aggregate = sum(v for _, v in child_values) / len(child_values)
                msg = f"CHANCE cập nhật kỳ vọng tạm thời={aggregate:.3f}."
            events.append(make_debug_event(
                "UPDATE", state, player, depth, path,
                msg,
                role=role,
                value=aggregate,
                current_move=move,
                child_values=list(child_values),
            ))

        final_value = max(v for _, v in child_values) if player == MAX_PLAYER else sum(v for _, v in child_values) / len(child_values)
        events.append(make_debug_event(
            "RETURN", state, player, depth, path,
            f"{role} trả value={final_value:.3f}.",
            role=role,
            value=final_value,
            child_values=list(child_values),
        ))
        return final_value

    recurse(board, root_player, 0, [])
    return events

class AdversarialSearchApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Buổi 14 - Tìm kiếm đối kháng")
        self.root.geometry("1240x760")
        self.root.minsize(1160, 700)

        self.algorithm_var = tk.StringVar(value="Minimax")
        self.opponent_var = tk.StringVar(value="Optimal")
        self.speed_var = tk.IntVar(value=700)
        self.debug_depth_var = tk.IntVar(value=3)
        self.board = empty_board()
        self.turn = MAX_PLAYER
        self.history = []
        self.last_result = None
        self.auto_job = None
        self.debug_events = []
        self.debug_index = 0

        self._build_style()
        self._build_ui()
        self.render()

    def _build_style(self):
        style = ttk.Style()
        try:
            style.theme_use("clam")
        except tk.TclError:
            pass
        style.configure("TFrame", background="#f5f7fb")
        style.configure("Panel.TFrame", background="white", relief="solid", borderwidth=1)
        style.configure("TLabel", background="#f5f7fb", font=("Segoe UI", 10))
        style.configure("Title.TLabel", font=("Segoe UI", 17, "bold"), foreground="#111827")
        style.configure("Subtitle.TLabel", font=("Segoe UI", 10), foreground="#4b5563")
        style.configure("PanelTitle.TLabel", background="white", font=("Segoe UI", 11, "bold"))
        style.configure("TButton", font=("Segoe UI", 10))

    def _build_ui(self):
        main = ttk.Frame(self.root, padding=16)
        main.pack(fill="both", expand=True)

        ttk.Label(main, text="Buổi 14 - Tìm kiếm đối kháng", style="Title.TLabel").pack(anchor="w")
        ttk.Label(
            main,
            text="Step-debug Minimax, Alpha-Beta Pruning và Expectimax trên Tic-Tac-Toe",
            style="Subtitle.TLabel",
        ).pack(anchor="w", pady=(2, 12))

        controls = ttk.Frame(main)
        controls.pack(fill="x", pady=(0, 12))

        ttk.Label(controls, text="Thuật toán").pack(side="left")
        algo_box = ttk.Combobox(
            controls,
            textvariable=self.algorithm_var,
            values=("Minimax", "Alpha-Beta", "Expectimax"),
            state="readonly",
            width=15,
        )
        algo_box.pack(side="left", padx=(6, 12))
        algo_box.bind("<<ComboboxSelected>>", lambda _event: self.debug_reset())

        ttk.Label(controls, text="Đối thủ O").pack(side="left")
        opponent_box = ttk.Combobox(
            controls,
            textvariable=self.opponent_var,
            values=("Optimal", "Random"),
            state="readonly",
            width=10,
        )
        opponent_box.pack(side="left", padx=(6, 12))

        ttk.Label(controls, text="Debug depth").pack(side="left")
        ttk.Spinbox(controls, from_=1, to=5, textvariable=self.debug_depth_var, width=4, command=self.debug_reset).pack(side="left", padx=(6, 12))

        ttk.Button(controls, text="Debug Reset", command=self.debug_reset).pack(side="left", padx=3)
        ttk.Button(controls, text="Debug Step", command=self.debug_step).pack(side="left", padx=3)
        ttk.Button(controls, text="AI Move", command=self.ai_move).pack(side="left", padx=3)
        ttk.Button(controls, text="Opponent Move", command=self.opponent_move).pack(side="left", padx=3)
        ttk.Button(controls, text="Reset", command=self.reset).pack(side="left", padx=3)
        ttk.Button(controls, text="Undo", command=self.undo).pack(side="left", padx=3)

        body = ttk.Frame(main)
        body.pack(fill="both", expand=True)

        left = ttk.Frame(body, style="Panel.TFrame", padding=12)
        left.pack(side="left", fill="both", expand=True, padx=(0, 10))
        right = ttk.Frame(body, style="Panel.TFrame", padding=12)
        right.pack(side="right", fill="both", expand=False)

        ttk.Label(left, text="Algorithm Debugger", style="PanelTitle.TLabel").pack(anchor="w")
        self.canvas = tk.Canvas(left, width=780, height=590, bg="white", highlightthickness=0)
        self.canvas.pack(fill="both", expand=True, pady=(10, 0))
        self.canvas.bind("<Button-1>", self.on_canvas_click)

        ttk.Label(right, text="Current Debug Event", style="PanelTitle.TLabel").pack(anchor="w")
        self.metrics_text = tk.Text(right, width=48, height=23, font=("Consolas", 10), bg="#ffffff", relief="flat")
        self.metrics_text.pack(fill="x", pady=(10, 10))
        self.metrics_text.configure(state="disabled")

        ttk.Label(right, text="Log", style="PanelTitle.TLabel").pack(anchor="w")
        self.log_text = tk.Text(right, width=48, height=16, font=("Consolas", 9), bg="#111827", fg="#e5e7eb", relief="flat")
        self.log_text.pack(fill="both", expand=True, pady=(10, 0))
        self.log_text.configure(state="disabled")

    def reset(self):
        self.stop_auto()
        self.board = empty_board()
        self.turn = MAX_PLAYER
        self.history = []
        self.last_result = None
        self.debug_events = []
        self.debug_index = 0
        self.clear_log()
        self.append_log("Reset game. X/MAX đi trước.")
        self.render()

    def debug_reset(self):
        depth = int(self.debug_depth_var.get())
        self.debug_events = build_debug_events(self.board, self.algorithm_var.get(), self.turn, depth)
        self.debug_index = 0
        self.append_log(f"Debug reset: {self.algorithm_var.get()}, depth={depth}, events={len(self.debug_events)}.")
        self.render()

    def debug_step(self):
        if not self.debug_events:
            self.debug_reset()
            return
        if self.debug_index < len(self.debug_events) - 1:
            self.debug_index += 1
        self.render()

    def current_debug_event(self):
        if not self.debug_events:
            return None
        return self.debug_events[self.debug_index]

    def on_canvas_click(self, event):
        if self.turn != MIN_PLAYER or winner(self.board) is not None:
            return
        move = self.cell_from_xy(event.x, event.y)
        if move is None or self.board[move] != EMPTY:
            return
        self.push_state()
        self.board = board_after(self.board, move, MIN_PLAYER)
        self.turn = MAX_PLAYER
        self.debug_events = []
        self.debug_index = 0
        self.append_log(f"Human/O đi {move_label(move)}.")
        self.last_result = None
        self.render()

    def cell_from_xy(self, x, y):
        start = 35
        cell = 105
        if not (start <= x <= start + 3 * cell and start + 40 <= y <= start + 3 * cell + 40):
            return None
        col = int((x - start) // cell)
        row = int((y - start - 40) // cell)
        if row < 0 or row > 2 or col < 0 or col > 2:
            return None
        return row * 3 + col

    def push_state(self):
        self.history.append((self.board, self.turn, self.last_result, list(self.debug_events), self.debug_index))

    def undo(self):
        if not self.history:
            return
        self.board, self.turn, self.last_result, self.debug_events, self.debug_index = self.history.pop()
        self.append_log("Undo một bước.")
        self.render()

    def ai_move(self):
        if self.turn != MAX_PLAYER or winner(self.board) is not None:
            return
        result = evaluate_with_algorithm(self.board, self.algorithm_var.get())
        self.last_result = result
        move = result["move"]
        if move is None:
            return
        self.push_state()
        self.board = board_after(self.board, move, MAX_PLAYER)
        self.turn = MIN_PLAYER
        self.debug_events = []
        self.debug_index = 0
        self.append_log(f"X/MAX dùng {self.algorithm_var.get()} chọn {move_label(move)} với value={result['value']:.3f}.")
        self.render()

    def opponent_move(self):
        if self.turn != MIN_PLAYER or winner(self.board) is not None:
            return
        moves = legal_moves(self.board)
        if not moves:
            return
        self.push_state()
        if self.opponent_var.get() == "Random":
            move = random.choice(moves)
            label = "Random"
        else:
            stats = {"nodes": 0, "pruned": 0, "trace": []}
            scored = [
                (move, minimax(board_after(self.board, move, MIN_PLAYER), MAX_PLAYER, 1, stats))
                for move in moves
            ]
            scored.sort(key=lambda item: (item[1], item[0]))
            move = scored[0][0]
            label = "Optimal/min"
        self.board = board_after(self.board, move, MIN_PLAYER)
        self.turn = MAX_PLAYER
        self.debug_events = []
        self.debug_index = 0
        self.append_log(f"O/MIN ({label}) đi {move_label(move)}.")
        self.last_result = None
        self.render()

    def auto_play(self):
        self.append_log("Auto Play đã được thay bằng Debug Step để tập trung trực quan hóa thuật toán.")

    def stop_auto(self):
        if self.auto_job is not None:
            self.root.after_cancel(self.auto_job)
            self.auto_job = None

    def make_preview(self):
        if winner(self.board) is None:
            return evaluate_with_algorithm(self.board, self.algorithm_var.get())
        if self.last_result is not None:
            return self.last_result
        return {
            "move": None,
            "value": utility(self.board),
            "candidates": [],
            "nodes": 0,
            "pruned": 0,
            "trace": [],
            "message": "Game đã kết thúc.",
        }

    def render(self):
        preview = self.make_preview()
        event = self.current_debug_event()
        self.canvas.delete("all")
        self.draw_board(preview)
        self.draw_debug_tree(preview, event)
        self.render_metrics(preview, event)

    def draw_board(self, preview):
        start = 35
        cell = 105
        self.canvas.create_text(start, 30, text="Current game board", anchor="w", font=("Segoe UI", 12, "bold"), fill="#111827")
        self.draw_mini_board(self.board, start, 70, cell, big=True)

        result = winner(self.board)
        if result == MAX_PLAYER:
            status = "X/MAX thắng"
        elif result == MIN_PLAYER:
            status = "O/MIN thắng"
        elif result == "Draw":
            status = "Hòa"
        else:
            status = f"Lượt hiện tại: {self.turn}"
        self.canvas.create_text(190, 420, text=status, font=("Segoe UI", 14, "bold"), fill="#111827")
        self.canvas.create_text(
            35, 455,
            text=f"Best root move: {move_label(preview['move']) if preview['move'] is not None else '-'} | value={preview['value']:.3f}",
            anchor="w",
            font=("Segoe UI", 10, "bold"),
            fill="#111827",
        )

    def draw_debug_tree(self, preview, event):
        x0 = 400
        self.canvas.create_text(x0, 30, text="Step-by-step recursion debugger", anchor="w", font=("Segoe UI", 12, "bold"), fill="#111827")

        self.canvas.create_rectangle(x0, 55, x0 + 330, 125, fill="#e0f2fe", outline="#111827", width=2)
        self.canvas.create_text(x0 + 12, 74, text=f"ROOT / {self.algorithm_var.get()}", anchor="w", font=("Segoe UI", 10, "bold"), fill="#111827")
        self.canvas.create_text(x0 + 12, 96, text=f"Candidates={len(preview.get('candidates', []))} | nodes={preview.get('nodes', 0)} | pruned={preview.get('pruned', 0)}", anchor="w", font=("Segoe UI", 9), fill="#374151")

        for index, (move, value) in enumerate(preview.get("candidates", [])[:6]):
            x = x0 + (index % 3) * 110
            y = 150 + (index // 3) * 56
            fill = "#bbf7d0" if move == preview.get("move") else "#f8fafc"
            self.canvas.create_rectangle(x, y, x + 98, y + 42, fill=fill, outline="#111827", width=2)
            self.canvas.create_text(x + 8, y + 14, text=f"X->{move_label(move)}", anchor="w", font=("Segoe UI", 8, "bold"), fill="#111827")
            self.canvas.create_text(x + 8, y + 31, text=f"v={value:.3f}", anchor="w", font=("Segoe UI", 8), fill="#374151")

        if event is None:
            self.canvas.create_rectangle(x0, 285, x0 + 330, 520, fill="#fff7ed", outline="#111827", width=2)
            self.canvas.create_text(x0 + 14, 310, text="Nhấn Debug Reset rồi Debug Step", anchor="w", font=("Segoe UI", 10, "bold"), fill="#111827")
            self.canvas.create_text(x0 + 14, 334, text="để xem thuật toán đi xuống/quay lên từng node.", anchor="w", font=("Segoe UI", 9), fill="#374151")
            return

        phase_color = {
            "ENTER": "#dbeafe",
            "DESCEND": "#fef3c7",
            "LEAF": "#e0e7ff",
            "UPDATE": "#dcfce7",
            "PRUNE": "#fecaca",
            "RETURN": "#ede9fe",
        }.get(event["phase"], "#f8fafc")
        self.canvas.create_rectangle(x0, 255, x0 + 330, 540, fill=phase_color, outline="#111827", width=2)
        header = f"{self.debug_index + 1}/{len(self.debug_events)}  {event['phase']}  depth={event['depth']}  role={event.get('role', event['player'])}"
        self.canvas.create_text(x0 + 12, 278, text=header, anchor="w", font=("Segoe UI", 9, "bold"), fill="#111827")
        self.canvas.create_text(x0 + 12, 300, text=format_path(event["path"])[:52], anchor="w", font=("Segoe UI", 8), fill="#374151")
        self.canvas.create_text(x0 + 12, 322, text=event["message"][:58], anchor="w", font=("Segoe UI", 8), fill="#111827")

        self.draw_mini_board(event["board"], x0 + 14, 350, 38, big=False)

        info_x = x0 + 145
        if "alpha" in event:
            self.canvas.create_text(info_x, 356, text=f"alpha={event['alpha']:.3f}", anchor="w", font=("Consolas", 9), fill="#111827")
        if "beta" in event:
            self.canvas.create_text(info_x, 376, text=f"beta ={event['beta']:.3f}", anchor="w", font=("Consolas", 9), fill="#111827")
        if "value" in event:
            self.canvas.create_text(info_x, 396, text=f"value={event['value']:.3f}", anchor="w", font=("Consolas", 9, "bold"), fill="#111827")
        if event.get("current_move") is not None:
            self.canvas.create_text(info_x, 416, text=f"move ={move_label(event['current_move'])}", anchor="w", font=("Consolas", 9), fill="#111827")

        child_values = event.get("child_values", [])
        if child_values:
            text = ", ".join(f"{move_label(m)}:{v:.1f}" for m, v in child_values[:5])
            self.canvas.create_text(x0 + 12, 492, text=f"returned: {text}", anchor="w", font=("Segoe UI", 8), fill="#374151")
        pruned = event.get("pruned_moves", [])
        if pruned:
            text = ", ".join(move_label(move) for move in pruned)
            self.canvas.create_text(x0 + 12, 514, text=f"pruned: {text}", anchor="w", font=("Segoe UI", 8, "bold"), fill="#b91c1c")

    def draw_mini_board(self, board, x, y, cell, big=False):
        font_size = 42 if big else 15
        self.canvas.create_rectangle(x - 6, y - 6, x + 3 * cell + 6, y + 3 * cell + 6, fill="#f8fafc", outline="#111827", width=2)
        for i in range(4):
            pos = i * cell
            self.canvas.create_line(x, y + pos, x + 3 * cell, y + pos, fill="#111827", width=2)
            self.canvas.create_line(x + pos, y, x + pos, y + 3 * cell, fill="#111827", width=2)
        for index, value in enumerate(board):
            row = index // 3
            col = index % 3
            cx = x + col * cell + cell / 2
            cy = y + row * cell + cell / 2
            if value == MAX_PLAYER:
                self.canvas.create_text(cx, cy, text="X", font=("Segoe UI", font_size, "bold"), fill="#2563eb")
            elif value == MIN_PLAYER:
                self.canvas.create_text(cx, cy, text="O", font=("Segoe UI", font_size, "bold"), fill="#dc2626")
            elif big:
                self.canvas.create_text(cx, cy, text=move_label(index), font=("Segoe UI", 8), fill="#9ca3af")

    def render_metrics(self, preview, event):
        candidate_lines = []
        for move, value in preview["candidates"]:
            mark = "*" if move == preview["move"] else " "
            candidate_lines.append(f"{mark} {move_label(move):>4}: {value:>7.3f}")

        lines = [
            f"Algorithm : {self.algorithm_var.get()}",
            f"Turn      : {self.turn}",
            f"Best move : {move_label(preview['move']) if preview['move'] is not None else '-'}",
            f"Value     : {preview['value']:.3f}",
            f"Nodes     : {preview['nodes']}",
            f"Pruned    : {preview['pruned']}",
            f"Winner    : {winner(self.board) or '-'}",
            "",
            "Root candidates:",
            *(candidate_lines or ["-"]),
            "",
            "Current debug event:",
        ]

        if event is None:
            lines.extend(["-", "Nhấn Debug Reset / Debug Step."])
        else:
            lines.extend([
                f"Index     : {self.debug_index + 1}/{len(self.debug_events)}",
                f"Phase     : {event['phase']}",
                f"Depth     : {event['depth']}",
                f"Role      : {event.get('role', event['player'])}",
                f"Path      : {format_path(event['path'])}",
                f"Board     : {short_board(event['board'])}",
                f"Message   : {event['message']}",
            ])
            if "alpha" in event:
                lines.append(f"Alpha     : {event['alpha']:.3f}")
            if "beta" in event:
                lines.append(f"Beta      : {event['beta']:.3f}")
            if "value" in event:
                lines.append(f"Value     : {event['value']:.3f}")
            if event.get("child_values"):
                returned = ", ".join(f"{move_label(m)}={v:.2f}" for m, v in event["child_values"])
                lines.append(f"Returned  : {returned}")
            if event.get("pruned_moves"):
                lines.append("Pruned    : " + ", ".join(move_label(m) for m in event["pruned_moves"]))

        lines.extend(["", "Ghi chú:", preview["message"]])

        self.metrics_text.configure(state="normal")
        self.metrics_text.delete("1.0", "end")
        self.metrics_text.insert("end", "\n".join(lines))
        self.metrics_text.configure(state="disabled")

    def append_log(self, text):
        self.log_text.configure(state="normal")
        self.log_text.insert("end", text + "\n")
        self.log_text.see("end")
        self.log_text.configure(state="disabled")

    def clear_log(self):
        self.log_text.configure(state="normal")
        self.log_text.delete("1.0", "end")
        self.log_text.configure(state="disabled")



if __name__ == "__main__":
    root = tk.Tk()
    app = AdversarialSearchApp(root)
    root.mainloop()
````
