```python
import tkinter as tk
import random

# =========================
# 8 Puzzle Game - Python Tkinter
# =========================
# 0 = ô trống

SIZE = 3

# Ma trận đích
GOAL_STATE = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]


class Puzzle8:
    def __init__(self, root):
        self.root = root
        self.root.title("8 Puzzle")

        self.board = [
            [1, 3, 5],
            [4, 6, 8],
            [2, 7, 0]
        ]

        self.buttons = []

        # Frame chứa puzzle
        self.frame = tk.Frame(root)
        self.frame.pack(pady=10)

        # Tạo giao diện
        for i in range(SIZE):
            numberButton = [] # actually 3 buttons in a row
            for j in range(SIZE):
                btn = tk.Button(
                    self.frame,
                    text="",
                    width=6,
                    height=3,
                    font=("Arial", 24),
                    command=lambda r=i, c=j: self.click_tile(r, c)
                )
                btn.grid(row=i, column=j, padx=2, pady=2)
                numberButton.append(btn)

            self.buttons.append(numberButton)

        # Frame điều khiển
        control_frame = tk.Frame(root)
        control_frame.pack(pady=10)

        self.update_ui()

    # =========================
    # Tìm vị trí ô trống
    # =========================
    def find_empty(self):
        for i in range(SIZE):
            for j in range(SIZE):
                if self.board[i][j] == 0: #cell nào bằng 0 thì là empty
                    return i, j

    # =========================
    # Cập nhật giao diện
    # =========================
    def update_ui(self):
        for i in range(SIZE):
            for j in range(SIZE):
                value = self.board[i][j]

                if value == 0:
                    self.buttons[i][j].config(text="", bg="lightgray")
                else:
                    self.buttons[i][j].config(
                        text=str(value),
                        bg="white"
                    )


    # =========================
    # Click ô để di chuyển
    # =========================
    def click_tile(self, r, c):
        empty_r, empty_c = self.find_empty()

        # Chỉ cho đổi nếu cạnh ô trống
        if abs(r - empty_r) + abs(c - empty_c) == 1:
            self.board[empty_r][empty_c], self.board[r][c] = \
                self.board[r][c], self.board[empty_r][empty_c]

            self.update_ui()


# =========================
# Main
# =========================
root = tk.Tk()
game = Puzzle8(root)
root.mainloop()
```
